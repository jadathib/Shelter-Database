```python
from dash import Dash, html, dcc, dash_table, Input, Output, callback_context
import plotly.express as px
import pandas as pd
from animal_shelter import AnimalShelter

# Create instance of the AnimalShelter class
shelter = AnimalShelter("myUserAdmin", "trinidad1")

# Initialize the Dash app
app = Dash(__name__)

# Create the layout
app.layout = html.Div([
    # Title
    html.Center(html.B(html.H1('SNHU CS-340 Dashboard'))),

    # Filter buttons
    html.Center(html.Div([
        html.Button('Cats', id='cat-button', n_clicks=0),
        html.Button('Dogs', id='dog-button', n_clicks=0),
        html.Button('Reset', id='reset-button', n_clicks=0)
    ])),

    # Data table
    dash_table.DataTable(
        id='datatable',
        columns=[
            {"name": i, "id": i} for i in ['animal_type', 'breed', 'color', 'age_upon_outcome']
        ],
        page_size=10
    ),

    # Graph
    dcc.Graph(id='data-graph')
])

# Callback to update dashboard
@app.callback(
    [Output('datatable', 'data'),
     Output('data-graph', 'figure')],
    [Input('cat-button', 'n_clicks'),
     Input('dog-button', 'n_clicks'),
     Input('reset-button', 'n_clicks')]
)
def update_dashboard(cat_clicks, dog_clicks, reset_clicks):
    ctx = callback_context
    if not ctx.triggered:
        query = {}
    else:
        button_id = ctx.triggered[0]['prop_id'].split('.')[0]
        if button_id == "cat-button":
            query = {"animal_type": "Cat"}
        elif button_id == "dog-button":
            query = {"animal_type": "Dog"}
        else:
            query = {}

    # Get data
    cursor = shelter.read_all(query)
    df = pd.DataFrame(list(cursor))

    # Create visualization
    if not df.empty:
        fig = px.bar(df['breed'].value_counts().head(10),
                     title='Top 10 Breeds')
    else:
        fig = px.bar(title='No Data Available')

    return df.to_dict('records'), fig

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
